```python
 # ========================================================
# 1 Case Study: Baseball
# 1.2 Data Preparation
# ========================================================

# ========================================================
# MODULE 11: Renaming Columns (.rename)
# ========================================================
# Our dataset currently has a column named "Team.name.2014".
#
# This name is:
# - Long and awkward to type
# - Easy to mistype (because of capital letters, dots, and numbers)
# - Not very clean for analysis
#
# Best practice in programming:
# Use short, clear, human-friendly names.
#
# So we want to rename "Team.name.2014" to simply "team".
# --------------------------------------------------------

# Step 1: The tool we need — .rename()
# --------------------------------------------------------
# pandas gives us a method called .rename().
#
# Syntax:
#   DataFrame.rename(columns={old_name: new_name}, inplace=True)
#
# Let’s break this down:
# - DataFrame.rename(...) → means: “use the rename tool
#   that belongs to a DataFrame.”
#   Since our DataFrame is called datapay, we write datapay.rename(...).
#
# - columns= → tells pandas we want to rename *columns*
#   (rename() could also rename rows, but we don’t need that).
#
# - {old_name: new_name} → this part is a dictionary.
#   A dictionary is a Python data structure for storing pairs of keys and values.
#   Each pair is written inside curly braces { }.
#   Example: {"cat": "meow", "dog": "bark"}
#   - Key = "cat", Value = "meow"
#   - Key = "dog", Value = "bark"
#
#   In our case:
#   {"Team.name.2014": "team"}
#   - Key = the current column name in the dataset
#   - Value = the new name we want
#
# - inplace=True → tells pandas to apply the change *directly*
#   to datapay (not just make a copy).
#
# So the full command is:
datapay.rename(columns={"Team.name.2014": "team"}, inplace=True)
 # --------------------------------------------------------
# Step 2: Verify the change
# --------------------------------------------------------
# Let’s check the first 10 column names to confirm
# that the rename worked.
#
# We use [:10] to slice the list of column names.
# Slicing means: “take items from index 0 up to (but not including) 10.”
# Remember: Python starts counting at 0.
print("First 10 column names after renaming:")
print(datapay.columns[:10])

# --------------------------------------------------------
# What you should see in the output:
# --------------------------------------------------------
# Index(['Unnamed: 0', 'team', 'p1998', 'p1999', 'p2000',
#        'p2001', 'p2002', 'p2003', 'p2004', 'p2005'],
#       dtype='object')
#
# Notice:
# - "Team.name.2014" has been replaced with "team"
# - "Unnamed: 0" is still there (this column came from the CSV
#   storing row numbers — later we’ll learn how to drop it).
#
# Why this matters:
# - Shorter names = less typing, fewer mistakes
# - Cleaner code = easier to read and share
# - A great example of how dictionaries (a core Python data structure)
#   are used to *map* one label into another
# ========================================================
```

```python
# ========================================================
# 1 Case Study: Baseball
# 1.2 Data Preparation
# ========================================================

# ========================================================
# MODULE 12: Reshaping the Data (wide → long, using melt)
# ========================================================

# --------------------------------------------------------
# Step 0: Where we are right now
# --------------------------------------------------------
# Our dataset "datapay" is in WIDE format.
# Wide means: one row has many years of data squished into it.
#
# Example:
#   team                 p1998   p1999   X1998   X1999   X1998.pct
#   Arizona Diamondbacks   31.6    70.5     65     100     0.401
#
# The YEAR is buried inside the column names ("p1998", "X1999").
# We want to pull it out into its own column called "year".
#
# --------------------------------------------------------
# What we want instead: TIDY DATA
# --------------------------------------------------------
# In tidy data:
# - Each COLUMN = one variable (team, year, payroll, wins, win_pct)
# - Each ROW = one observation (one team in one year)
#
# Example of tidy format:
#   team                 year   payroll   win_num   win_pct
#   Arizona Diamondbacks 1998    31.6       65        0.401
#   Arizona Diamondbacks 1999    70.5      100        0.617
#
# --------------------------------------------------------
# Step 1: Melt payroll columns
# --------------------------------------------------------
# pandas gives us a function called pd.melt().
# Melt = take many columns and "melt" them into two columns:
#   - one for the old column name
#   - one for the values inside
#
# Let's do it for payroll (columns named p1998, p1999, …).
payroll = pd.melt(
    datapay,
    id_vars=["team"],   # columns to keep fixed (identifier variable)
    value_vars=[col for col in datapay.columns if col.startswith("p")],
    var_name="year",    # new column to hold the old column names
    value_name="payroll"  # new column to hold the numbers
)


# --------------------------------------------------------
# Deep breakdown:
# --------------------------------------------------------
# id_vars=["team"]
# - "id_vars" means: which columns we are NOT melting.
# - We keep "team" fixed so each row still shows the team.
# - ["team"] is a LIST with one item.
#   A list in Python is written with [ ] brackets, items separated by commas.
#
# value_vars=[col for col in datapay.columns if col.startswith("p")]
# - This is a LIST COMPREHENSION.
# - It builds a list of column names automatically.
# - The structure is: [expression for item in sequence if condition]
# - Here:
#     col = each column name in datapay.columns
#     if col.startswith("p") → only keep it if it begins with "p"
# - So this produces: ["p1998", "p1999", ..., "p2014"].
#
# var_name="year"
# - The old column names (like "p1998") will go into a new column
#   called "year".
#
# value_name="payroll"
# - The values (like 31.6, 70.5) will go into a new column called "payroll".

# --------------------------------------------------------
# Step 1.1: Clean up the "year" column
# --------------------------------------------------------
# Right now, "year" looks like "p1998".
# We want just the number 1998.
#
# payroll["year"] → selects the "year" column
# .str.replace("p","") → removes the "p"
# .astype(int) → converts text into an integer
payroll["year"] = payroll["year"].str.replace("p", "").astype(int)

# --------------------------------------------------------
# Step 2: Melt win_num columns
# --------------------------------------------------------
# Wins are in columns like X1998, X1999, ...
# Same idea, but keep only columns starting with "X"
# and exclude ones ending with ".pct".
win_num = pd.melt(
    datapay,
    id_vars=["team"],
    value_vars=[col for col in datapay.columns if col.startswith("X") and ".pct" not in col],
    var_name="year",
    value_name="win_num"
)

# Clean up "year" (remove X, make integer)
win_num["year"] = win_num["year"].str.replace("X", "").astype(int)

# --------------------------------------------------------
# Step 3: Melt win_pct columns
# --------------------------------------------------------
# Win percentage columns look like X1998.pct, X1999.pct, ...
# We melt those too.
win_pct = pd.melt(
    datapay,
    id_vars=["team"],
    value_vars=[col for col in datapay.columns if col.endswith(".pct")],
    var_name="year",
    value_name="win_pct"
)

# Clean up "year": remove X and .pct
win_pct["year"] = win_pct["year"].str.replace("X", "").str.replace(".pct", "").astype(int)

# --------------------------------------------------------
# Step 4: Merge the three pieces
# --------------------------------------------------------
# Now we glue payroll, win_num, and win_pct into one DataFrame.
# The tool is .merge().
#
# Syntax:
#   table1.merge(table2, on=["col1","col2"])
#
# Logic:
# - Take table1 and table2
# - Match rows where col1 and col2 are the same
# - Join them together
#
# We merge twice: first payroll with win_num, then with win_pct.
baseball_long = (
    payroll
    .merge(win_num, on=["team", "year"])
    .merge(win_pct, on=["team", "year"])
)

# --------------------------------------------------------
# Step 5: Preview the result
# --------------------------------------------------------
print("Reshaped dataset (first 5 rows):")
print(baseball_long.head())

# Expected output (example):
#   team                  year   payroll   win_num   win_pct
#   Arizona Diamondbacks  1998    31.6       65        0.401
#   Atlanta Braves        1998    61.7       89        0.617
#
# Each row now = one team in one year.
# This is TIDY DATA in Python.

# --------------------------------------------------------
# Step 6: Save the tidy dataset
# --------------------------------------------------------
# We save the reshaped dataset as baseball.csv so we can
# reload it later without repeating all this reshaping.
#
# .to_csv() = write a DataFrame to a CSV file.
# index=False = don’t write row numbers as a separate column.
baseball_long.to_csv("baseball.csv", index=False)

print("Saved tidy dataset as baseball.csv")

```

```python
# ========================================================
# 2 Exploratory Data Analysis (EDA)
# 2.1 Part I: Analyze Aggregated Variables
# ========================================================

# ========================================================
# MODULE 13: Input the Clean Data (baseball.csv)
# ========================================================

# --------------------------------------------------------
# Step 0: Why Exploratory Data Analysis (EDA)?
# --------------------------------------------------------
# EDA is always the first step after we tidy data.
# The goal is to:
#   - Understand what variables exist
#   - Summarize them (averages, ranges, spreads)
#   - Explore relationships between them
#
# Think of EDA as the "getting to know you" phase with the dataset.
# Before we build models, we need to know the basics.
# --------------------------------------------------------

# --------------------------------------------------------
# Step 1: Load the clean dataset
# --------------------------------------------------------
# In Module 12, we reshaped the messy wide-format data into tidy form
# and saved it as "baseball.csv" in our *current working directory*.
#
# That means the file lives right here where Colab is running,
# not inside any subfolder.
#
# To load it back into Python, we use the pandas function pd.read_csv().
# Syntax:
#   pd.read_csv("filename.csv")
#
# The input "filename.csv" must exactly match the name of the file
# (and the path, if it’s inside a folder).
# Since we saved it as "baseball.csv" — not "data/baseball.csv" —
# we must write "baseball.csv".
baseball = pd.read_csv("baseball.csv")

# --------------------------------------------------------
# Step 2: Confirm columns
# --------------------------------------------------------
# Let’s confirm that the dataset has the expected columns.
print("Column names:", baseball.columns.tolist())

# Expected output:
#   ['team', 'year', 'payroll', 'win_num', 'win_pct']
# Each of these is now a clear, tidy variable.

# --------------------------------------------------------
# Step 3: Inspect structure
# --------------------------------------------------------
# To look deeper into the structure, we use the .info() method.
# .info() tells us:
#   - How many rows and columns there are
#   - The data type of each column
#   - Whether there are missing values
print("\nStructure of the DataFrame:")
baseball.info()

# --------------------------------------------------------
# Step 4: Breaking down the .info() output
# --------------------------------------------------------
# Here’s what you’ll typically see:
#
# <class 'pandas.core.frame.DataFrame'>
# RangeIndex: 510 entries, 0 to 509
# Data columns (total 5 columns):
#   Column   Non-Null Count   Dtype
# 0 team     510 non-null     object
# 1 year     510 non-null     int64
# 2 payroll  510 non-null     float64
# 3 win_num  510 non-null     int64
# 4 win_pct  510 non-null     float64
#
# Let’s decode:
# - "RangeIndex: 510 entries, 0 to 509"
#   → The dataset has 510 rows (one row per team-year).
# - "team" column → 510 values, type = object (text, e.g., "Yankees").
# - "year" column → 510 values, type = int64 (whole numbers, like 1998).
# - "payroll" column → 510 values, type = float64 (decimals, millions).
# - "win_num" column → 510 values, type = int64 (whole numbers, wins).
# - "win_pct" column → 510 values, type = float64 (decimals, between 0 and 1).
#
# At the bottom, pandas summarizes the counts of each data type
# and the memory usage in RAM.

# --------------------------------------------------------
# Step 5: Why this matters
# --------------------------------------------------------
# Compared to before (wide data):
# - Each ROW is now one team in one year.
# - Each COLUMN is one variable (team, year, payroll, win_num, win_pct).
# - No variables are trapped inside column names.
#
# This is exactly what "tidy data" means.
# With tidy data confirmed, we are now ready to start EDA.
# ========================================================
```

```python
# ========================================================
# 2 Exploratory Data Analysis (EDA)
# 2.1 Part I: Analyze Aggregated Variables
# ========================================================

# ========================================================
# MODULE 14: Aggregate Payroll and Win Percentage
# ========================================================

# --------------------------------------------------------
# Step 0: Why are we doing this?
# --------------------------------------------------------
# Right now, our dataset has one row for *each team in each year*.
# That’s useful for detailed analysis, but sometimes we want
# the "big picture." For example:
#
#   - Over 17 years, how much money did each team spend?
#   - On average, how often did each team win?
#
# These are *aggregated questions* — questions that require
# combining many rows of data into one summary per team.
#
# In data science, this process is called:
#
#   → AGGREGATION
#
# Aggregation = grouping data + summarizing groups.
#
# Why do we care?
# - It reveals long-term patterns (not just year-to-year noise).
# - It’s used everywhere: business (sales by customer),
#   healthcare (average blood pressure per patient),
#   sports (total payroll per team).
# --------------------------------------------------------

# --------------------------------------------------------
# Step 1: Group the data by team
# --------------------------------------------------------
# pandas gives us a tool called .groupby().
#
# Syntax:
#   baseball.groupby("team")
#
# - baseball → the name of our DataFrame
# - .groupby("team") → tells pandas:
#     "Split the rows into groups, one group for each team name."
#
# IMPORTANT:
#   At this point, no math is done. The rows are just
#   *organized into buckets* behind the scenes.
#
# Think of it like sorting papers into folders:
# - All "Atlanta Braves" rows go in one folder,
# - All "Boston Red Sox" rows go in another,
# - and so on, for all 30 teams.
grouped = baseball.groupby("team")

# --------------------------------------------------------
# Step 2: Summarize each group
# --------------------------------------------------------
# Once the data is grouped, we can calculate statistics
# for each folder (each team).
#
# In this case:
#   - payroll_total = the sum of payrolls across all years
#   - win_pct_ave   = the average win percentage across all years
#
# pandas has a method called .agg() (short for aggregate).
# Inside .agg(), we tell pandas:
#   "For this column, apply this function."
#
# Syntax:
#   grouped.agg(new_column_name = ("old_column", "function"))
#
# Example:
#   ("payroll", "sum") means "sum up all payroll values"
#   ("win_pct", "mean") means "take the average win percentage"

data_agg = grouped.agg(
    payroll_total = ("payroll", "sum"),   # add up payrolls
    win_pct_ave   = ("win_pct", "mean")   # take average win percentage
)

# --------------------------------------------------------
# Step 3: Adjust the units
# --------------------------------------------------------
# The payroll numbers are currently in MILLIONS of dollars.
# Example: 2000 means $2,000 million = $2 billion.
#
# To make these totals easier to read, we convert them
# from millions to billions by dividing by 1000.
#
# - data_agg["payroll_total"] selects the payroll_total column.
# - /1000 divides each value by 1000.
data_agg["payroll_total"] = data_agg["payroll_total"] / 1000

# --------------------------------------------------------
# Step 4: Reset the index
# --------------------------------------------------------
# Right now, "team" is being used as the index (row labels).
# If we want "team" back as a regular column,
# we use .reset_index().
data_agg = data_agg.reset_index()

# --------------------------------------------------------
# Step 5: Preview the new table
# --------------------------------------------------------
print("Aggregated payroll (in billions) and average win percentage by team:")
print(data_agg.head())

# --------------------------------------------------------
# What the output means:
# --------------------------------------------------------
# Each row now represents ONE TEAM across ALL YEARS.
#
# Columns:
# - team          → the team name
# - payroll_total → total payroll (in billions)
# - win_pct_ave   → average win percentage
#
# Example row:
#   team          payroll_total   win_pct_ave
#   Atlanta Braves     1.52          0.563
#
# Translation:
#   Across 17 years, the Braves spent $1.52 billion
#   and won about 56.3% of their games.
#
# --------------------------------------------------------
# Why this matters for data science:
# --------------------------------------------------------
# Aggregation is one of the most important steps in analysis.
# It lets us:
#   - Collapse thousands of rows into digestible summaries.
#   - Compare categories (teams, customers, hospitals, etc.).
#   - Ask questions about long-term averages and totals.
#
# This is the bridge between raw data and insight.
# Without aggregation, we’d be stuck staring at rows of numbers.
# With aggregation, we can see patterns that actually matter.
# ========================================================

```

```python
mean_payroll_total = data_agg["payroll_total"].mean()

# Let’s decode the syntax:
# - data_agg → our DataFrame (the whole table).
# - ["payroll_total"] → the square brackets [ ] select ONE column by name.
#   This gives us just that column as a Series (think of it like a vector).
# - .mean → the dot (.) means "use a method that belongs to this object."
#   Every Series has methods like .mean(), .median(), etc.
# - () → parentheses are how we actually CALL the method.
#
# So: data_agg["payroll_total"].mean()
# means "from the table data_agg, take the payroll_total column,
# then compute its mean."

mean_win_pct_ave = data_agg["win_pct_ave"].mean()

print("Mean payroll_total (billions):", mean_payroll_total)
print("Mean win_pct_ave:", mean_win_pct_ave)


# --------------------------------------------------------
# Step 2: The MEDIAN
# --------------------------------------------------------
# The median is the middle value when numbers are sorted.
# Unlike the mean, the median is not affected by extreme values.
#
# By hand:
#   Example: [10, 15, 20] → median = 15
#   Example: [10, 20, 30, 100] → median = (20 + 30)/2 = 25
#
# In Python, the pattern is the same as before:
#   DataFrame["column"].median()

median_payroll_total = data_agg["payroll_total"].median()
median_win_pct_ave   = data_agg["win_pct_ave"].median()

print("\nMedian payroll_total (billions):", median_payroll_total)
print("Median win_pct_ave:", median_win_pct_ave)


# --------------------------------------------------------
# Step 3: QUANTILES
# --------------------------------------------------------
# Quantiles split the data into cut points.
# Quartiles (quarters) are most common:
#   - 25% quantile (Q1): one quarter of teams fall below this
#   - 50% quantile (Q2): the median
#   - 75% quantile (Q3): three quarters of teams fall below this
#
# By hand:
#   Imagine sorting payrolls smallest to largest.
#   - The 25% quantile is the value 1/4 of the way down.
#   - The 75% quantile is 3/4 of the way down.
#
# In Python:
#   DataFrame["column"].quantile(decimal)
#   - 0.25 → 25%
#   - 0.75 → 75%

q25_payroll_total = data_agg["payroll_total"].quantile(0.25)
q75_payroll_total = data_agg["payroll_total"].quantile(0.75)

q25_win_pct_ave = data_agg["win_pct_ave"].quantile(0.25)
q75_win_pct_ave = data_agg["win_pct_ave"].quantile(0.75)

print("\n25th percentile payroll_total:", q25_payroll_total)
print("75th percentile payroll_total:", q75_payroll_total)
print("25th percentile win_pct_ave:", q25_win_pct_ave)
print("75th percentile win_pct_ave:", q75_win_pct_ave)


# --------------------------------------------------------
# Step 4: FULL SUMMARY with .describe()
# --------------------------------------------------------
# Instead of running mean(), median(), quantile() one by one,
# pandas has a shortcut: .describe()
#
# Syntax:
#   DataFrame[["col1","col2"]].describe()
#
# Notice:
# - [["col1","col2"]] uses double square brackets because
#   we’re selecting MULTIPLE columns (a list of strings).
# - .describe() gives us:
#     count → number of teams
#     mean  → arithmetic average
#     std   → standard deviation (spread)
#     min   → smallest value
#     25%, 50%, 75% → quartiles
#     max   → largest value

print("\nSummary of aggregates:")
print(data_agg[["payroll_total", "win_pct_ave"]].describe())


# --------------------------------------------------------
# Step 5: Why all these measures matter
# --------------------------------------------------------
# Each measure tells us something different:
#
# - Mean → the balance point, but sensitive to outliers.
# - Median → the middle team, resistant to extreme values.
# - Quantiles → how spread out the values are (low vs. high teams).
# - .describe() → a quick one-stop summary of everything.
#
# In our baseball data:
# - Payroll totals range from ~0.7 to ~2.8 billion.
#   This huge range shows financial inequality between teams.
# - Average win percentages cluster near 0.5.
#   That makes sense: every game has one winner and one loser,
#   so league-wide the average hovers around 50%.
#
# Data science takeaway:
# Before modeling, we ALWAYS stop to compute these descriptive statistics.
# They help us see the center, the spread, and whether there are
# extreme values that might distort our analysis.
# ========================================================
```

```python
max_payroll_index = data_agg["payroll_total"].idxmax()


# --------------------------------------------------------
# Step 2: Using .loc[] to get the actual team name
# --------------------------------------------------------
# Now that we know the row index, we need the TEAM NAME.
#
# pandas has .loc[] for this.
# Syntax:
#   dataframe.loc[row_index, "column_name"]
#
# - .loc → means "look inside the table by labels."
# - row_index → which row do we want? (we got that from .idxmax()).
# - "column_name" → which column’s value do we want?
#
# So:
#   data_agg.loc[max_payroll_index, "team"]
# means:
#   Go to that row, and grab the team name.
team_max_payroll = data_agg.loc[max_payroll_index, "team"]

print("Team with maximum payroll_total:", team_max_payroll)


# --------------------------------------------------------
# Step 3: Repeat for the minimum payroll
# --------------------------------------------------------
# Now we do the same thing, but instead of .idxmax(),
# we use .idxmin().
#
# .idxmin() gives the row index of the smallest value.
min_payroll_index = data_agg["payroll_total"].idxmin()

team_min_payroll = data_agg.loc[min_payroll_index, "team"]

print("Team with minimum payroll_total:", team_min_payroll)
```

```python
# ========================================================
# 2 Exploratory Data Analysis (EDA)
# 2.1 Part I: Analyze Aggregated Variables
# ==================================================

# ========================================================
# MODULE 18 (Part 3): Normal Variable — Distribution of win_pct_ave
# ========================================================

# Step 1: Calculate mean and standard deviation
# --------------------------------------------------------
# We start by calculating the mean and standard deviation
# of win_pct_ave, which is each team’s average win percentage
# across 17 seasons.
#
# Syntax reminder:
#   data_agg["win_pct_ave"].mean()
#
# - data_agg is our DataFrame (like a table in memory).
# - ["win_pct_ave"] uses square brackets to select a single column.
#   When you do this, pandas gives you back a Series, which is
#   basically a one-dimensional array of numbers.
# - .mean() is a method (a built-in function attached to objects).
#   The dot (.) means “use the mean function that belongs to this Series.”
# - Python calculates the arithmetic average:
#   add up all the values and divide by how many there are.
#
# Similarly, .std() gives the standard deviation, which measures
# how spread out the values are around the mean.

mean_win = data_agg["win_pct_ave"].mean()
sd_win   = data_agg["win_pct_ave"].std()

print("Mean of win_pct_ave:", mean_win)
print("Standard deviation of win_pct_ave:", sd_win)

# Interpretation:
# The mean is about 0.50, meaning the typical team wins half its games.
# The standard deviation is about 0.038, so most teams cluster within
# about 4 percentage points of 0.50. This makes sense in MLB:
# there’s parity, and teams hover around a 50% win rate.


# Step 2: Create values for a smooth normal curve
# --------------------------------------------------------
# To see if our data looks normal (bell-shaped),
# we’ll overlay a normal curve on top of the histogram.
#
# To do that, we need x-values (from 0.0 to 1.0) and
# the corresponding y-values from the normal probability
# density function (PDF).
#
# We use NumPy for x-values.
# NumPy is Python’s main library for numerical work.
# It’s optimized for fast math and built on the ndarray,
# which is a powerful n-dimensional array structure.
#
# Syntax:
#   np.arange(start, stop, step)
# - np is the nickname for NumPy (we write import numpy as np).
# - arange means “array range.”
# - start=0, stop=1, step=0.001 → numbers from 0.000 to 0.999
#   in steps of 0.001. That gives us 1000 equally spaced points.

import numpy as np
from scipy.stats import norm   # SciPy has statistical distributions

x = np.arange(0, 1, 0.001)

# Next we compute the y-values using the normal PDF.
# Syntax:
#   norm.pdf(x, loc=mean, scale=sd)
#
# - norm is the normal distribution object.
# - pdf stands for probability density function.
# - x is the array of points where we want the curve.
# - loc sets the center (the mean).
# - scale sets the spread (the standard deviation).
#
# This gives us the y-values for the bell curve
# that matches our data’s mean and SD.

y = norm.pdf(x, loc=mean_win, scale=sd_win)


# Step 3: Plot the histogram and the normal curve
# --------------------------------------------------------
# Now we compare the actual data distribution to the theoretical normal.
#
# For the histogram:
#   sns.histplot(data, bins=10, stat="density", color="red", alpha=0.6)
#
# - data = data_agg["win_pct_ave"], the column we’re plotting.
# - bins=10 → split the range into 10 intervals.
# - stat="density" rescales the bars so the total area = 1.
#   This way, the histogram and the normal curve share the same scale.
# - color="red" fills the bars with red.
# - alpha=0.6 makes the bars semi-transparent.
#
# For the curve:
#   plt.plot(x, y, color="blue", linewidth=3)
#
# - x and y are the arrays we created earlier.
# - color="blue" makes the line blue.
# - linewidth=3 makes it thicker so it stands out.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(data_agg["win_pct_ave"], bins=10, stat="density", color="red", alpha=0.6)
plt.plot(x, y, color="blue", linewidth=3)

plt.title("Histogram of Average Win Percentage with Normal Curve")
plt.xlabel("Average Win Percentage (win_pct_ave)")
plt.ylabel("Density")

plt.show()

# Interpretation:
# The red bars show the observed data distribution.
# The smooth blue curve shows the normal distribution with the same mean and SD.
# They overlap closely, which tells us win_pct_ave is roughly normal.


# Step 4: Apply the empirical rule
# --------------------------------------------------------
# For data that’s normal, the empirical rule says:
# - About 68% of values fall within 1 SD of the mean.
# - About 95% fall within 2 SDs.
# - About 99.7% fall within 3 SDs.
#
# We can compute the first two ranges explicitly:

lower_68 = mean_win - sd_win
upper_68 = mean_win + sd_win

lower_95 = mean_win - 2*sd_win
upper_95 = mean_win + 2*sd_win

print("\n68% interval:", (lower_68, upper_68))
print("95% interval:", (lower_95, upper_95))

# Interpretation:
# - 68% of teams fall between about 0.462 and 0.538.
# - 95% fall between about 0.425 and 0.575.
#
# These intervals describe what’s “typical” in baseball wins.
# A team consistently above 0.575 would be considered exceptional,
# while one consistently below 0.425 would be unusually poor.


# Step 5: Why this matters beyond baseball
# --------------------------------------------------------
# Many algorithms in data science and machine learning,
# such as linear regression, logistic regression, and neural networks,
# rely on assumptions about input variables.
# If variables are approximately normal, then mean and SD
# are sufficient to describe them, and standardization (z-scores)
# works well to put features on the same scale.
#
# In sports analytics, this matters for fairness comparisons.
# By knowing that win_pct_ave is roughly normal,
# we can model it with fewer numbers and predict ranges more reliably.
```
